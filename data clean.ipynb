```python
import pandas as pd
data = pd.read_csv("emissions_c.csv")
```

```python
record_num = int(data.describe().ix[0,0])
record_num 
```

```python
year_emi ={}
for i in range(1999, 2019, 1):
    year_emi[i] = [0,0,0]

for i in range(record_num ):
    row = data.ix[i,:]
    air = int(row['air_total_emission_kg'])
    water = int(row['water_emission_kg'])
    land = int(row['land_emission_kg'])
    year_emi[row['report_year']][0]+= air
    year_emi[row['report_year']][1]+= water
    year_emi[row['report_year']][2]+= land
    
```

```python
print(year_emi)
```

```python
air_c=[]
water_c =[]
land_c =[]
for i in range(1999, 2019, 1):
    air_c.append(year_emi[i][0])
    water_c.append(year_emi[i][1])
    land_c.append(year_emi[i][2])
year_emi={}
```

```python
year_emi['air'] = air_c
year_emi['water'] = water_c
year_emi['land'] = land_c
year_emi
```

```python
years=[]
for i in range(1999, 2019, 1):
    years.append(i)
df = pd.DataFrame(year_emi,columns=['air','water','land'], index = years)
print(df)
```
